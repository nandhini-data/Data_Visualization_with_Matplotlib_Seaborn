## Matplotlib Basics 

**Matplotlib** is an open source Python library built for data visualization that let you produce a wide variety of highly customizable charts and graphs.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.plot([0,5,10,20,30,40,50])
```

## Compatible Data Types

Matplotlib can plot many **datatypes**, including base python sequences,NumPy Arrays,Pandas Series and DataFrames

```python
x=[]
for i in range(10):
    y=i *2
    x.append(y)
print(x)
```

```python
plt.plot(x)
```

```python
plt.plot(pd.Series(x))
```

```python
data =[x *11 for x in range(10)]
data
```

```python
labels =[f'{x}$' for x in range(10) ]
```

```python
labels
```

```python
plt.plot(labels,data)

plt.show()
```

```python
df = pd.DataFrame(data,labels).assign(y = lambda x:x[0]/2)
df
```

```python
plt.plot(df)
plt.show()
```

## Plotting Methods

Matplotlib has two plotting methods or interfaces:

* **PyPlot API** - Charts are created with the plot() function,and modified with additional functions.
* **Object-Oriented** - Charts are created by defining a plot object,and modified using figure and axis methods.

```python
fig,ax = plt.subplots()  #create the figure and axis
ax.plot(x)          #plots x

fig.suptitle("Main Title") # Add a title to the figure and axis
ax.set_title("Chart")
```

```python
plt.plot(data)
plt.title("Title")
plt.show()
```

```python
fig =plt.figure()
ax =fig.add_subplot()

ax.plot(data)

ax.set_title("Title")
fig.suptitle("Overall title")
```

## Plotting DataFrames

When **plotting dataframes** using the object-oriented interface,Matplotlib will use the index as the x-axis and plot each column as a seperate series by default.

```python
data_dict ={"sales":[10,20,30,40],"profits":[2,4,5,1],"Date":["2022-01-01","2022-02-01","2022-03-01","2022-04-01"]}
```

```python
df=pd.DataFrame(data_dict,columns=["sales","profits"],index=data_dict["Date"])
```

```python
df
```

```python
fig,ax = plt.subplots()
ax.plot(df)
ax.set_title("sales and profits")
```

```python
fig,ax= plt.subplots()
ax.plot(df.index,df["sales"],ls ="--")
ax.set_title("sales")
```

```python
fig,ax= plt.subplots()
ax.plot(df.index,df["profits"],ls ="-.")
ax.set_title("sales")
```
