# <center> Data Visualization</center>

 Data Visualization allow you to **bring your data to life**<br>
 * The human brain is built to interpret raw data as meaningless numbers and noise.
 * We need clear patterns and visual cues to help us quickly make sense of complex information.

## The 3 key questions

The 3 key questions are a great way to help choose the right visual.<br>
1. What **type of data** are you working with?
    * Time -series
    * Categorical
    * Numeric
    * Hierarchical
    
2. What do you want to **communicate**?
    * Comparison
    * Composition
    * Distribution
    * Relationship
    
3. Who is the **end user** and what do they need?
    * Analyst
    * Manager
    * Executive
    * General public

## Chart Formatting


**Chart Formatting** should be used to eliminate noise and facilitate understanding.<br>

**Pro tips**<br>
* Remove the chart border and gridline
* Format the axis label clearly
* Add context with the chart title
* Create a visual order
* Make sure the story is clear

## Storytelling

Descriptive title and data labels can be used to **tell a clear story** within your visuals.<br>

**Pro tip**
* Leverage the title to guide the audience toward specific insights.
* Insert text and shape directly inside the chart.
* Use data labels and annotation to draw attention to the main data points
* Use color strategically.
